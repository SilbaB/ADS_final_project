**importing the necessary libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**importing the dataset**

```python
data=pd.read_csv("Student_performance_data _.csv")

```

```python
data.head()
```

```python
data.shape
```

```python
data.dtypes
```

```python
data['GradeClass']=data['GradeClass'].astype(int)
data.dtypes
```

```python
data=data.drop(columns=['StudentID'],axis=1)
```

```python
data.isna().sum()
```

**Data visualization**

```python
data.describe()
```

```python
for columns in data:
    plt.figure(figsize=(10,5))
    sns.countplot(x=columns,data=data)
    plt.title(f'Distribution of {columns}')
    plt.show()
```

```python
for columns in data:
    plt.style.use('ggplot')
    plt.figure(figsize=(7,5))
    plt.hist(x=columns,data=data,color='cyan',edgecolor='black')
    plt.xlabel(f'Distribution of {columns}')
plt.show()
```

```python
correlation=data.corr()
```

```python

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, fmt=".02f", cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
